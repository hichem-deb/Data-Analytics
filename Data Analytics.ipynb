```
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.metrics import f1_score

```

```
total = Data.isnull().sum().sort_values(ascending=False)
print(total)
```

```
W1_mean = Data["W1"].mean()
W1_mean= round(W1_mean)
W2_mean = Data["W2"].mean()
W2_mean= round(W2_mean)
W3_mean = Data["W3"].mean()
W3_mean= round(W3_mean)
Employed_mean = Data["Employed"].mean()
Employed_mean= round(Employed_mean)
```

```
Data["W1"].fillna(W1_mean, inplace=True)
Data["W2"].fillna(W2_mean, inplace=True)
Data["W3"].fillna(W3_mean, inplace=True)
Data["Employed"].fillna(Employed_mean, inplace=True)
```

```
new_total = Data.isnull().sum().sort_values(ascending=False)
print(new_total)
```

```
New_Data= pd.read_excel('New_Data.xls')
New_Data.head()
```

```
Data_Corr= New_Data.iloc[:,2:-1]
Data_Corr.head()
```

```
Data_Corr.shape
```

```
correlation = Data_Corr.corr()
sns.heatmap(correlation, vmax=1, square=True,cmap='YlGnBu')
plt.title('Correlation between different features')
```

```
columns = np.full((correlation.shape[0],), True, dtype=bool)
for i in range(correlation.shape[0]):
    for j in range(i+1, correlation.shape[0]):
        if correlation.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
```

```
selected_columns = Data_Corr.columns[columns]
selected_columns.shape
```

```
selected_columns
```

```
X=New_Data[selected_columns]
y=New_Data['Class']
```

```
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size = 0.33,random_state=0)
```

```
clf=RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```
y_pred
```

```
y_true=np.array(y_test)
y_true
```

```
f1_score(y_true, y_pred, average='micro')
```
